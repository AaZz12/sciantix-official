```python
number = 6
directory = f"Gold_{number}"
file_path_gold = f"{directory}/output.txt"
directory = f"ROM 5_{number}"
file_path_rom5 = f"{directory}/output.txt"
directory = f"ROM 10_{number}"
file_path_rom10 = f"{directory}/output.txt"
```

```python

fig, (ax1) = plt.subplots(1,  figsize=(5,5)) 
data = pd.read_table(file_path_gold)
ax1.plot(data['Time (h)'],data['Xe in grain (at/m3)'], label='Gold', linewidth=1.5)
data = pd.read_table("ROM 5_6/output.txt")
ax1.plot(data['Time (h)'],data['Xe in grain (at/m3)'], label='ROM sphere 5 modes', linewidth=1.5)
data = pd.read_table("ROM 10_6/output.txt")
ax1.plot(data['Time (h)'],data['Xe in grain (at/m3)'], '--',label='ROM sphere 10 modes', linewidth=1.5)
ax1.set_xlabel('Time (h)',fontsize=14)
ax1.set_ylabel('Temperature (K)',fontsize=14)
ax1.tick_params(axis="y", labelsize=13)
ax1.tick_params(axis="x", labelsize=13)
ax1.legend()
#plt.savefig(os.path.join('plot','Temperature.pdf'), bbox_inches='tight')
```
